```python
data = pd.DataFrame(columns=["hotel_url", "author", "date", "rating", "title", "text"])
```

```python
for line in file.readlines():
    line = eval(line)
    line.pop("property_dict")
    line["text"] = str(line["text"]).encode('utf-16', 'surrogatepass').decode('utf-16')
    lk = pd.DataFrame(line,
                    columns=["hotel_url", "author", "date", "rating", "title", "text"],
                    index=pd.RangeIndex(0,1))
    data = pd.concat([data, lk], ignore_index=True)
```
