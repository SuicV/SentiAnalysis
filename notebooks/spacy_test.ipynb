```python
import time
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
d1 = nlp("This hotel is outstanding. The staff is friendly and attentive. It’s a great location that is quiet yet close to the subway to get anywhere you need to go quickly and it feels very safe. The price is right too! Highly recommend and would stay again. Thanks so much to the entire staff for making our stay amazing!")
```

```python
d1
```

```python
for sentence in d1.sents:
    for token in sentence:
        print(token.text, token.pos_, token.dep_)
    print("=======")
```

```python
displacy.render(d1)
```
